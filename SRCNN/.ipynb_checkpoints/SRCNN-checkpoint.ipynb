```python
import PIL.Image as pilimg
import numpy as np
import matplotlib.pyplot as plt
import random
import os
```

```python
img = pilimg.open("SR_dataset/291/000t1.bmp").convert("L")
h = img.size[0]
w = img.size[1]

train_y = np.array(img)
img = img.resize((int(h/2),int(w/2)))
test = np.array(img)
img = img.resize((h,w))
train_x = np.array(img)


plt.imshow(train_y)
```

```python
def draw_image(img1,img2,title,subTitle1,sub2):
    fig = plt.figure()
    fig.suptitle(title)
    ax1 = fig.add_subplot(1,2,1)
    ax1.imshow(img1,cmap=plt.cm.gray)
    ax1.set_title(subTitle1)
    ax2 = fig.add_subplot(1,2,2)
    ax2.imshow(img2,cmap = plt.cm.gray)
    ax2.set_title(sub2)
    
    plt.show()

draw_image(train_y,train_x,"image","origin","noisy")
```

```python
def random_crop(X,Y,crop_size = (32,32)):
    h = X.shape[0]
    w = X.shape[1]
    offset_h = random.randint(0, h-crop_size[0])
    offset_w = random.randint(0, w-crop_size[1])
    x = X[offset_h: offset_h + crop_size[0], offset_w: offset_w + crop_size[1]]
    y = Y[offset_h: offset_h + crop_size[0], offset_w: offset_w + crop_size[1]]
    
    return x,y

x,y = random_crop(train_x, train_y)
draw_image(y,x,"image","origin","noisy")
```

```python
path = "./SR_dataset/291"
def read_image(path):
    file_list = os.listdir(path)
    #print(file_list)
    # x - noisy data, y - origin data
    train_x = []
    train_y = []
    for file in file_list:
        if file != "Thumbs.db":
            img = pilimg.open(path+"/"+file).convert("L")
        else:
            continue
        h = img.size[0]
        w = img.size[1]

        origin_img = np.array(img)
        img = img.resize((int(h/2),int(w/2)))
        img = img.resize((h,w))
        noise_img = np.array(img)
        x,y =  random_crop(noise_img,origin_img)
        #print(x.shape,y.shape)
        train_x.append(x)
        train_y.append(y)
    return np.array(train_x), np.array(train_y)    
    
train_x, train_y = read_image(path)

print("test")
draw_image(train_x[3],train_y[3],"image","noisy","origin")
```

```python
print(train_x.shape,train_y.shape)
train_y = np.reshape(train_y,(291,32,32,1))
print(train_y.shape)
train_y ,train_x = train_y / 255 , train_x /255
```

```python
from tensorflow.keras import backend as K

def psnr(y_true,y_pred):
    max_pixel = 1.0
    return 10.0 / 2.303 * K.log((max_pixel ** 2) / 
                        (K.mean(K.square(y_pred-y_true))))
```

```python
'''
define model
3-layered CNN
'''
import tensorflow as tf
from tensorflow.keras.callbacks import TensorBoard

model = tf.keras.models.Sequential([
    tf.keras.layers.Reshape((32,32,1)),
    tf.keras.layers.Conv2D(8, (3,3), padding='same', activation = 'relu'),
    tf.keras.layers.Conv2D(16, (3,3), padding='same', activation = 'relu'),
    tf.keras.layers.Conv2D(1,(3,3), padding='same')
])
model.compile(optimizer='adam',
            loss = 'mean_squared_error',
             metrics = [psnr])
tensorboard = TensorBoard(log_dir = './log', histogram_freq = 0,
                         write_graph = True, write_images= False)

model.fit(train_x, train_y,epochs=100,batch_size=128,callbacks = [tensorboard])
```
